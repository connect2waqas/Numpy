```python
# layer = [0,1,2]
# rows = [0,1,2]
# column = [1,1,2]

# data_3d[layer,rows,column]
# data_3d
```

```python
# layers = [0, 1, 2]
# rows   = [1, 2, 0]
# cols   = [2, 3, 0]
# data_3d[layer,rows,cols]

layers = [0,1]
rows = [1,3]
cols = [3,3]
data_3d[layers,rows,cols]
data_3d
```

```python
layers = [2,1,0]
rows = [2,3,2]
cols = [2,3,2]
print(data_3d[layers,rows,cols])
data_3d[[2,1,0],[2,3,2],[2,3,2]]
```
