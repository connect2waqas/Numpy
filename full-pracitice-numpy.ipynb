Array Operations:

```python
k = np.arange(12).reshape(3,4)
l = np.arange(12,24).reshape(3,4)
```

```python
k + l
k % l
k - l
k / l
k.nbytes == l.nbytes
k.itemsize == l.itemsize
k.dtype == l.dtype
```

```python
k * 2 * l
```

```python
3 * l
```
