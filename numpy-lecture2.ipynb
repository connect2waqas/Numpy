Sorting of Array:

```python
a = np.random.randint(1,100,15)
b = np.random.randint(1,100,24).reshape(6,4)
print(a)
print(b)
```

```python
"Rows wise sorting of array: (by default they sorts the values row wise axis= 0)"
print(np.sort(a))
print(np.sort(b))
# columns wise sorting array:
print(np.sort(a,axis=0))
print(np.sort(b,axis=0))

"What if we have to sort in reversed order:"

print(np.sort(a)[::-1])
print(np.sort(b) [::-1])

"We can do this through the sorted"
c = sorted(a)
d = sorted(a)
print(c)
```

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)
print(c)
print(d)
```
